```python
import pandas as pd
df = pd.read_csv("dftest_27sep2024.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(df['Strike_call'], df['Implied_Volatility_Call'], marker='o', color='b', linestyle='-')  # Create a line plot with markers
plt.title('Implied Volatility vs. Strike Price')
plt.xlabel('Strike_call')
plt.ylabel('Implied_Volatility_Call')
plt.grid(True)  # Show grid
plt.show()

plt.plot(df['Strike_put'], df['Implied_Volatility_Put'], marker='o', color='r', linestyle='-')  # Create a line plot with markers
plt.title('Implied Volatility vs. Strike Price')
plt.xlabel('Strike_put')
plt.ylabel('Implied_Volatility_Put')
plt.grid(True)  # Show grid
plt.show()

# df['Strike_put'] df['Implied_Volatility_Put'] 'Strike_put' 'Implied_Volatility_Put'
```
